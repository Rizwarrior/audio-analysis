```python
import mido
import os

def extract_drum_timestamps_mido(midi_file_path, output_txt_path="drum_timestamps.txt"):
    """
    Extract timestamps for kicks and snares from a MIDI file using mido library.
    
    Args:
        midi_file_path (str): Path to the MIDI file
        output_txt_path (str): Path for the output text file
    """
    
    try:
        # Load the MIDI file
        mid = mido.MidiFile(midi_file_path)
    except Exception as e:
        print(f"Error loading MIDI file: {e}")
        return
    
    # General MIDI drum mapping
    drum_mapping = {
        35: "Acoustic Bass Drum",
        36: "Bass Drum 1",          # Main kick
        38: "Acoustic Snare",       # Main snare
        40: "Electric Snare",       # Alternative snare
        42: "Closed Hi-Hat",
        46: "Open Hi-Hat",
        49: "Crash Cymbal 1",
        51: "Ride Cymbal 1"
    }
    
    # Focus on kicks and snares
    kick_notes = [35, 36]  # Bass drum notes
    snare_notes = [38, 40]  # Snare drum notes
    
    all_timestamps = []
    kick_timestamps = []
    snare_timestamps = []
    
    # Process each track
    for i, track in enumerate(mid.tracks):
        print(f'Processing track {i}: {track.name}')
        
        current_time = 0.0
        
        for msg in track:
            # Convert delta time to absolute time (in seconds)
            current_time += mido.tick2second(msg.time, mid.ticks_per_beat, 500000)  # 500000 = 120 BPM
            
            # Process note_on messages for drum channel (channel 9, which is index 9)
            if msg.type == 'note_on' and msg.channel == 9 and msg.velocity > 0:
                note_number = msg.note
                velocity = msg.velocity
                
                # Store all drum events
                drum_name = drum_mapping.get(note_number, f"Drum {note_number}")
                all_timestamps.append({
                    'time': current_time,
                    'note': note_number,
                    'drum': drum_name,
                    'velocity': velocity
                })
                
                # Categorize kicks and snares
                if note_number in kick_notes:
                    kick_timestamps.append({
                        'time': current_time,
                        'note': note_number,
                        'velocity': velocity
                    })
                elif note_number in snare_notes:
                    snare_timestamps.append({
                        'time': current_time,
                        'note': note_number,
                        'velocity': velocity
                    })
    
    # Sort timestamps
    all_timestamps.sort(key=lambda x: x['time'])
    kick_timestamps.sort(key=lambda x: x['time'])
    snare_timestamps.sort(key=lambda x: x['time'])
    
    # Write to text file
    with open(output_txt_path, 'w') as f:
        f.write("DRUM TRANSCRIPTION TIMESTAMPS\n")
        f.write("=" * 50 + "\n\n")
        
        f.write(f"Total drum events: {len(all_timestamps)}\n")
        f.write(f"Kick drums: {len(kick_timestamps)}\n")
        f.write(f"Snare drums: {len(snare_timestamps)}\n\n")
        
        # Write kick timestamps
        f.write("KICK DRUM TIMESTAMPS (seconds):\n")
        f.write("-" * 30 + "\n")
        for i, kick in enumerate(kick_timestamps, 1):
            f.write(f"{i:3d}. {kick['time']:8.3f}s (note {kick['note']}, velocity {kick['velocity']})\n")
        
        f.write("\n")
        
        # Write snare timestamps
        f.write("SNARE DRUM TIMESTAMPS (seconds):\n")
        f.write("-" * 30 + "\n")
        for i, snare in enumerate(snare_timestamps, 1):
            f.write(f"{i:3d}. {snare['time']:8.3f}s (note {snare['note']}, velocity {snare['velocity']})\n")
        
        f.write("\n")
        
        # Write all drum events for reference
        f.write("ALL DRUM EVENTS:\n")
        f.write("-" * 20 + "\n")
        for event in all_timestamps:
            f.write(f"{event['time']:8.3f}s - {event['drum']} (note {event['note']}, vel {event['velocity']})\n")
    
    # Print summary to console
    print(f"\n Results saved to: {output_txt_path}")
    print(f" Total drum events found: {len(all_timestamps)}")
    print(f" Kick drums: {len(kick_timestamps)}")
    print(f" Snare drums: {len(snare_timestamps)}")
    
    # Print first few timestamps for immediate viewing
    if kick_timestamps:
        print(f"\n KICK TIMESTAMPS (first 10):")
        for i, kick in enumerate(kick_timestamps[:10], 1):
            print(f"   {i:2d}. {kick['time']:8.3f}s")
    
    if snare_timestamps:
        print(f"\n SNARE TIMESTAMPS (first 10):")
        for i, snare in enumerate(snare_timestamps[:10], 1):
            print(f"   {i:2d}. {snare['time']:8.3f}s")
    
    return {
        'all_events': all_timestamps,
        'kicks': kick_timestamps,
        'snares': snare_timestamps
    }

# Example usage
if __name__ == "__main__":
    midi_file = "jazz_in_the_maybach.mp3.midi"  # Replace with actual output file name
    
    if os.path.exists(midi_file):
        results = extract_drum_timestamps_mido(midi_file)
    else:
        print(f"MIDI file '{midi_file}' not found.")
        print("Make sure to:")
        print("1. Run your Magenta transcription command first")
        print("2. Update the 'midi_file' variable with the correct output filename")
        
        # Scan for MIDI files
        midi_files = [f for f in os.listdir('.') if f.endswith('.mid')]
        if midi_files:
            print(f"\nFound these MIDI files in current directory:")
            for i, f in enumerate(midi_files, 1):
                print(f"  {i}. {f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import os
import matplotlib
matplotlib.use('Agg')  # Use non-interactive backend

# Fix for NumPy compatibility issues with pretty_midi
if not hasattr(np, 'int'):
    np.int = int
    np.float = float
    np.bool = bool

import pretty_midi

def extract_drum_timestamps(midi_file_path):
    """Extract kick and snare timestamps from MIDI file"""
    try:
        midi_data = pretty_midi.PrettyMIDI(midi_file_path)
    except Exception as e:
        print(f"Error loading MIDI file: {e}")
        return [], []
    
    kick_notes = [35, 36]  # Bass drum notes
    snare_notes = [38, 40]  # Snare drum notes
    
    kick_timestamps = []
    snare_timestamps = []
    
    for instrument in midi_data.instruments:
        if instrument.is_drum:
            for note in instrument.notes:
                if note.pitch in kick_notes:
                    kick_timestamps.append(note.start)
                elif note.pitch in snare_notes:
                    snare_timestamps.append(note.start)
    
    return sorted(kick_timestamps), sorted(snare_timestamps)

def create_drum_impulse_track(timestamps, time_array, amplitude=1.0):
    """Create impulse track array for drum timestamps"""
    impulse_track = np.zeros(len(time_array))
    
    for timestamp in timestamps:
        # Find closest time index
        idx = np.argmin(np.abs(time_array - timestamp))
        if 0 <= idx < len(impulse_track):
            impulse_track[idx] = amplitude
    
    return impulse_track

def visualize_drums_simple(audio_file, midi_file, output_image="drum_viz_simple.png"):
    """
    Create a simple, robust drum visualization that avoids cell block limits
    """
    print("=== SIMPLE DRUM VISUALIZATION ===")
    print("Loading and processing audio...")
    
    try:
        # Load audio with aggressive resampling and limiting
        target_sr = 16000  # Even lower sample rate
        y, sr = librosa.load(audio_file, sr=target_sr)
        duration = len(y) / sr
        
        # Limit to maximum 10k points for visualization
        max_viz_points = 10000
        if len(y) > max_viz_points:
            step = len(y) // max_viz_points
            y_viz = y[::step]
        else:
            y_viz = y
        
        # Create time array for visualization
        time_viz = np.linspace(0, duration, len(y_viz))
        
        print(f"Audio: {duration:.1f}s, {sr}Hz, Viz points: {len(y_viz)}")
        
    except Exception as e:
        print(f"Error loading audio: {e}")
        return
    
    print("Extracting drum timestamps...")
    kick_times, snare_times = extract_drum_timestamps(midi_file)
    print(f"Found {len(kick_times)} kicks, {len(snare_times)} snares")
    
    # Filter valid timestamps
    kick_valid = [t for t in kick_times if 0 <= t <= duration]
    snare_valid = [t for t in snare_times if 0 <= t <= duration]
    
    # Create drum impulse tracks
    kick_track = create_drum_impulse_track(kick_valid, time_viz, amplitude=1.0)
    snare_track = create_drum_impulse_track(snare_valid, time_viz, amplitude=1.0)
    
    print("Creating visualization...")
    
    # Clear everything and use the simplest possible approach
    plt.close('all')
    
    # Create figure with very simple layout
    fig = plt.figure(figsize=(15, 8), facecolor='white', dpi=100)
    
    # Create subplots using simple approach
    ax1 = fig.add_subplot(3, 1, 1)  # Waveform
    ax2 = fig.add_subplot(3, 1, 2)  # Kicks  
    ax3 = fig.add_subplot(3, 1, 3)  # Snares
    
    # Plot 1: Waveform (simple line plot only)
    ax1.plot(time_viz, y_viz, 'b-', linewidth=0.5, alpha=0.8)
    ax1.set_title(f'Audio: {os.path.basename(audio_file)} ({len(kick_valid)} kicks, {len(snare_valid)} snares)')
    ax1.set_ylabel('Amplitude')
    ax1.grid(True, alpha=0.3)
    ax1.set_xlim(0, duration)
    
    # Plot 2: Kick track (simple line plot)
    ax2.plot(time_viz, kick_track, 'r-', linewidth=1, label=f'Kicks ({len(kick_valid)})')
    ax2.fill_between(time_viz, kick_track, alpha=0.6, color='red')
    ax2.set_ylabel('Kick')
    ax2.set_ylim(0, 1.2)
    ax2.grid(True, alpha=0.3)
    ax2.set_xlim(0, duration)
    
    # Plot 3: Snare track (simple line plot)
    ax3.plot(time_viz, snare_track, 'orange', linewidth=1, label=f'Snares ({len(snare_valid)})')
    ax3.fill_between(time_viz, snare_track, alpha=0.6, color='orange')
    ax3.set_ylabel('Snare')
    ax3.set_xlabel('Time (seconds)')
    ax3.set_ylim(0, 1.2)
    ax3.grid(True, alpha=0.3)
    ax3.set_xlim(0, duration)
    
    # Add stats text
    stats_text = f"Duration: {duration:.1f}s | Kicks: {len(kick_valid)} | Snares: {len(snare_valid)}"
    if len(kick_valid) > 1:
        avg_kick_interval = np.mean(np.diff(kick_valid))
        stats_text += f" | Kick BPM: {60/avg_kick_interval:.0f}"
    
    ax1.text(0.02, 0.98, stats_text, transform=ax1.transAxes, 
             bbox=dict(boxstyle='round', facecolor='lightgray', alpha=0.8),
             verticalalignment='top', fontsize=9)
    
    # Simple layout adjustment
    fig.tight_layout()
    
    # Save with basic settings
    try:
        fig.savefig(output_image, dpi=150, facecolor='white', bbox_inches='tight')
        print(f" Saved: {output_image}")
    except Exception as e:
        print(f" Save error: {e}")
        # Try most basic save
        try:
            fig.savefig(output_image, facecolor='white')
            print(f" Saved (basic): {output_image}")
        except Exception as e2:
            print(f" Basic save failed: {e2}")
    
    plt.close(fig)
    return kick_valid, snare_valid

def create_comparison_chart(kick_times, snare_times, duration):
    """Create a simple timing comparison chart"""
    print("Creating timing comparison...")
    
    plt.close('all')
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 6), facecolor='white')
    
    # Create time grid for comparison
    time_grid = np.arange(0, duration, 0.1)  # 100ms resolution
    
    # Plot kick timing
    if kick_times:
        ax1.scatter(kick_times, [1]*len(kick_times), c='red', s=20, alpha=0.8, marker='|')
        ax1.set_ylabel('Kicks')
        ax1.set_ylim(0.5, 1.5)
        ax1.grid(True, alpha=0.3)
        ax1.set_title(f'Drum Timing Comparison ({len(kick_times)} kicks, {len(snare_times)} snares)')
    
    # Plot snare timing  
    if snare_times:
        ax2.scatter(snare_times, [1]*len(snare_times), c='orange', s=20, alpha=0.8, marker='|')
        ax2.set_ylabel('Snares')
        ax2.set_ylim(0.5, 1.5)
        ax2.grid(True, alpha=0.3)
        ax2.set_xlabel('Time (seconds)')
    
    plt.tight_layout()
    
    try:
        plt.savefig('drum_timing_comparison.png', dpi=150, facecolor='white')
        print(" Saved: drum_timing_comparison.png")
    except Exception as e:
        print(f" Comparison save error: {e}")
    
    plt.close(fig)

def analyze_drum_timing(kick_times, snare_times):
    """Simple timing analysis"""
    print("\n DRUM TIMING ANALYSIS")
    print("=" * 40)
    
    print(f" Totals:")
    print(f"   Kicks: {len(kick_times)}")
    print(f"   Snares: {len(snare_times)}")
    print(f"   Total hits: {len(kick_times) + len(snare_times)}")
    
    if len(kick_times) > 1:
        kick_intervals = np.diff(kick_times)
        print(f" Kick Analysis:")
        print(f"   Avg interval: {np.mean(kick_intervals):.3f}s")
        print(f"   Min interval: {np.min(kick_intervals):.3f}s")
        print(f"   Max interval: {np.max(kick_intervals):.3f}s")
        print(f"   Estimated BPM: {60/np.mean(kick_intervals):.1f}")
    
    if len(snare_times) > 1:
        snare_intervals = np.diff(snare_times)
        print(f" Snare Analysis:")
        print(f"   Avg interval: {np.mean(snare_intervals):.3f}s")
        print(f"   Min interval: {np.min(snare_intervals):.3f}s")  
        print(f"   Max interval: {np.max(snare_intervals):.3f}s")
        print(f"   Estimated BPM: {60/np.mean(snare_intervals):.1f}")

# Main execution
if __name__ == "__main__":
    audio_file = "dnb_alone_drums.wav"
    midi_file = "dnb_alone_drums.wav.midi"
    
    print(" DRUM VISUALIZATION TOOL")
    print("=" * 50)
    
    if os.path.exists(audio_file) and os.path.exists(midi_file):
        print(f"Processing: {audio_file}")
        print(f"MIDI file: {midi_file}")
        
        # Create main visualization
        kick_times, snare_times = visualize_drums_simple(audio_file, midi_file)
        
        if kick_times is not None:
            # Create comparison chart
            duration = None
            try:
                y, sr = librosa.load(audio_file, sr=16000)
                duration = len(y) / sr
                create_comparison_chart(kick_times, snare_times, duration)
            except:
                print("  Could not create comparison chart")
            
            # Analyze timing
            analyze_drum_timing(kick_times, snare_times)
            
            print("\n Visualization complete!")
            print(" Check for output files:")
            print("   - drum_viz_simple.png")
            print("   - drum_timing_comparison.png")
        else:
            print(" Visualization failed")
            
    else:
        print(" Files not found:")
        print(f"   Audio: {audio_file} ({'' if os.path.exists(audio_file) else ''})")
        print(f"   MIDI: {midi_file} ({'' if os.path.exists(midi_file) else ''})")
        
        # List available files
        print("\n Available files:")
        for f in os.listdir('.'):
            if f.endswith(('.wav', '.mp3', '.flac', '.mid', '.midi')):
                print(f"   {f}")
```
